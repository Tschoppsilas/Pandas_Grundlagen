```python
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/kimfetti/Videos/master/Pandas_Tips/data/pet_data.csv")
```

# Drop blank rows

```python
df.dropna(how='all',inplace=True)
```

```python
df.info()
```

# Find duplicate rows

```python
df[df.duplicated()]
```

```python
df[df.duplicated(keep=False)] #Keep=False muss sein, sonst erkenne ich nicht ob es duplikate sind oder nicht, nur das programm weiss es.
```

# Drop duplicate rows

```python
df.drop_duplicates()
```

```python
no_dupes = df.drop_duplicates()
```

```python
no_dupes[no_dupes.duplicated(keep=False)]
```

# Drop duplicates based on specific column(s)

```python
df.head()
```

```python
df.name.value_counts()
```

```python
df.drop_duplicates(subset='name') #entfertn nur duplikate aus der spalte name
```

```python
df.drop_duplicates(subset='name').name.value_counts()
```

## duplicate auf mehrere Spalten entfernen

```python
df.drop_duplicates(subset=['name', 'pet_type'])
```

```python
df.drop_duplicates(subset=['name', 'pet_type'])[['name', 'pet_type']].value_counts()
```

# Specify which duplicate to keep

```python
df.drop_duplicates(subset='name') #behält immer die erste Zeile mit einem einzigartigen Wert und löscht danach die doppelten
```

```python
df.drop_duplicates(subset='name', keep='last')#behält immer die letzte Zeile mit einem einzigartigen Wert und löscht somit die vorherigen
```

```python
df.drop_duplicates(subset='name', keep=False) #behält nur die, welche wirklich einzigartig sind. also sobald etwas doppelt ist, wird es automatisch gelöscht
```
