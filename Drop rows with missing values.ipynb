# Drop rows with missings in specific column(s)

```python
df.head()
```

```python
df.dropna(subset=['name']) #löscht alle zeilen bei denen der name NAN ist
```

```python
df.dropna(subset=['name']).info()
```

```python
df.dropna(subset=['name', 'pet_type']) #löscht alle zeilen bei denen der name oder die Tierart NAN ist
```

```python
df.dropna(subset=['name', 'pet_type'], how='all') #löscht wo name und Tierart NAN sind
```

```python
df.dropna(subset=['name', 'pet_type'], how='all').info()
```

# Drop columns with missing values

```python
df.dropna(axis=1) #führt dropna in den spalten durch
```

```python
df.dropna(axis=1, how='all')
```

## Set threshold for number of missings

```python
missing_threshold = df.shape[0]*0.5
missing_threshold
```

```python
df.dropna(axis=1, thresh=missing_threshold) #Löscht spalten mit mehr als thresholdwerte =NAN
```

```python
df.isna().sum()
```
