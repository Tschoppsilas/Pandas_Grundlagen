```python
import pandas as pd
```

```python
df = pd.read_csv("customers.csv", index_col = "ID")
```

```python
df.head()
```

# Spalten umbenennen

```python
df.rename(columns = {"First Name": "first_name"})
```

# Zeilen umbenennen

```python
df.rename(index={4576: "4576_Sophia"})
df.head()
```

# Dauerhafte Ã„nderungen

```python
df.head() #Man sieht es ist nicht permanent
```

```python
df.rename(columns = {"First Name": "first_name"}, inplace = True) #hier wird es permanent
```

```python
df.head()
```

```python
df.rename(
    index={
        4576: "4576_Sophia",
        9488 : "9488_Evelyn"
    },
    columns = {
        "First Name": "first_name",
        "Second Name": "second_name",
        "Phone": "phone",
        "LTV": "livetime_valie"
    }, inplace = True)
```

```python
df.head()
```
